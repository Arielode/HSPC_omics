```r
colors_border=c("#aa8439", "#992129", "#1a488c", "#7e6ba1", "grey" )
w$pvalue = NULL
w2 <- w %>% spread(clone, O2E)
w2 <- column_to_rownames(w2, "clus")
w2 <- w2[, c("clone1", "clone5", "clone9")]
w2 <- as.data.frame(t(w2))
w2[nrow(w2)+1, ] <- 1
w2[nrow(w2)+1, ] <- 0
w2[nrow(w2)+1, ] <- 3
w2 <- w2[c(nrow(w2), nrow(w2)-1, 1:(nrow(w2)-2)), ]
w2 <- w2[, c("HSC/MPP", "Myeloid", "Mk/Erythroid", "Lymphoid")]
```

```r
options(repr.plot.width=10, repr.plot.height=7)
radarchart( w2  , axistype=1, seg = 3, pty=16, 
            #custom polygon
            pcol=c("#D8261C", "#2C68AB", "#6B5298", 
            "grey") ,  plwd=2 , plty=2,
            #custom the grid
            cglcol="grey", cglty=1, axislabcol="grey", 
            caxislabels=seq(0,3,1),
            cglwd=0.1,
            #custom labels
            vlcex=1, calcex=1, maxmin = T
)
legend(x=1.8, y=1, legend = c(rownames(w2)[3:(nrow(w2)-1)], "Ctrl"), bty = "n", pch=20 , 
       col=c("#D8261C", "#2C68AB", "#6B5298",  "grey") , text.col = "black", cex=1, pt.cex=2)
text(x= 0.05 , y= 0.75 , "*")
text(x= 0.05 , y= -0.62 , "*")
text(x= 0.05 , y= -0.9 , "***")
```
